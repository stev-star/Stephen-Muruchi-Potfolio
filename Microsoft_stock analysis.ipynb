```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=[10,5])
plt.plot(df['Close'],'b')
plt.grid()
plt.xlabel('Year',fontsize=15)
plt.ylabel('Stock price',fontsize=15)
plt.title('Microsoft Stocks',fontsize=15)
plt.show()
```

```python
df['diff']=df['Open']-df['Close']
```

```python
df.tail()
```

```python
daily_close=df[['Adj Close']]
daily_pct_chng=daily_close.pct_change()
daily_pct_chng.fillna(0,inplace=True)
print(daily_pct_chng)
```

```python
daily_pct_chng.hist(bins=40)
plt.show()
print(daily_pct_chng.describe())
```

```python
adj_close_px=df['Adj Close']
df['50']=adj_close_px.rolling(window=50).mean()
df['100']=adj_close_px.rolling(window=100).mean()

plt.figure(figsize=[10,5])
plt.plot(df['Adj Close'],label='Adj close')
plt.plot(df['50'],label='50')
plt.plot(df['100'],label='100')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.title('Simple Moving Average',fontsize=15)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[10,5])
plt.plot(adj_close_px,label='Adj Close')
plt.plot(adj_close_px.ewm(span=50,adjust=False).mean(),label='50')
plt.plot(adj_close_px.ewm(span=100,adjust=False).mean(),label='100')
plt.title('Exponential Moving Average',fontsize=15)
plt.ylabel('Stock Price')
plt.xlabel('Year')
plt.grid()

plt.legend()
plt.show()
```

```python
min_periods=75
vol=daily_pct_chng.rolling(min_periods).std()*np.sqrt(min_periods)
plt.figure(figsize=[10,5])
plt.plot(vol)
plt.title('Volatility of Microsoft Stock',fontsize=15)
plt.xlabel('Year',fontsize=12)
plt.ylabel('stock price',fontsize=12)
plt.show()
```
